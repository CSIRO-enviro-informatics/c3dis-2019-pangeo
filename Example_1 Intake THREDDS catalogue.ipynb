```python
import intake
cat = intake.open_thredds_cat('http://thredds.aodn.org.au/thredds/catalog/IMOS/SRS/OC/gridded/aqua/P1D/catalog.xml')
```

Pick out a month - the \[''\]() syntax is a little ugly, but it works to return Intake catalogue objects and traverse the THREDDS server heirarchy

```python
month=cat['2007']()['02']()
month.url
```

Print the files available within the month

```python
for k in month:
    print(k)
```

Open a NetCDF file from thredds using the intake catalog ".to_dask()" function - it actually returns an XArray Dataset, the naming is an overhang from intake being developed originally for use with dask.

```python
ds=month['A.P1D.20070201T000000Z.aust.K_490.nc'].to_dask()
print(ds)
print('\n')
print(ds.K_490)
```
